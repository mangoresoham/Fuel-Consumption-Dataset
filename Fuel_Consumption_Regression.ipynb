# Encoding

```python
# Drop irrelevant columns (MAKE, MODEL, VEHICLECLASS, MODELYEAR)
new_df = df.drop(columns=['MAKE', 'MODEL', 'VEHICLECLASS', 'MODELYEAR','FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB_MPG'])

# Apply One-Hot Encoding to categorical features (TRANSMISSION, FUELTYPE)
new_df = pd.get_dummies(new_df, columns=['TRANSMISSION', 'FUELTYPE'], drop_first=True)

# Display the resulting DataFrame
new_df.head()
```

```python
# After One-Hot Encoding, convert boolean values to 0 and 1
new_df = new_df.astype(int)

# Display the resulting DataFrame
new_df.head()
```

```python
new_df.shape
```

# Standardization

```python
new_df.head()
```

```python
# List of remaining numerical features after dropping irrelevant ones
numerical_features = ['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS']
```

```python
# Apply standardization to the numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
new_df[numerical_features] = scaler.fit_transform(new_df[numerical_features])
```

```python
new_df.head()
```

# Splitting Data

```python
# Assuming new_df is a Pandas DataFrame
X = new_df.drop(columns=['CO2EMISSIONS'])  # Keep X as DataFrame
y = new_df['CO2EMISSIONS']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the model
lr_model = LinearRegression()

# Fit the model
lr_model.fit(X_train, y_train)

# Predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluation metrics
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = mean_squared_error(y_test, y_pred_lr, squared=False)
r2_lr = r2_score(y_test, y_pred_lr)
rss_lr = np.sum((y_test - y_pred_lr) ** 2)

print(f"Linear Regression:")
print(f"MAE: {mae_lr}")
print(f"MSE: {mse_lr}")
print(f"RSS: {rss_lr}")
print(f"RMSE: {rmse_lr}")
print(f"R^2: {r2_lr}")

```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluation metrics
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
r2_rf = r2_score(y_test, y_pred_rf)
rss_rf = np.sum((y_test - y_pred_rf) ** 2)

print(f"Random Forest Regressor:")
print(f"MAE: {mae_rf}")
print(f"MSE: {mse_rf}")
print(f"RMSE: {rmse_rf}")
print(f"R^2: {r2_rf}")
print(f"RSS: {rss_rf}")
```

# XG Boost

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the model
xgb_model = XGBRegressor(n_estimators=100, random_state=42)

# Fit the model
xgb_model.fit(X_train, y_train)

# Predictions
y_pred_xgb = xgb_model.predict(X_test)

# Evaluation metrics
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
rmse_xgb = mean_squared_error(y_test, y_pred_xgb, squared=False)
r2_xgb = r2_score(y_test, y_pred_xgb)
rss_xgb = np.sum((y_test - y_pred_xgb) ** 2)

print(f"XGBoost Regressor:")
print(f"MAE: {mae_xgb}")
print(f"MSE: {mse_xgb}")
print(f"RSS: {rss_xgb}")
print(f"RMSE: {rmse_xgb}")
print(f"R^2: {r2_xgb}")
```

# Support Vector Regressor

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the model
svr_model = SVR(kernel='rbf')

# Fit the model
svr_model.fit(X_train, y_train)

# Predictions
y_pred_svr = svr_model.predict(X_test)

# Evaluation metrics
mae_svr = mean_absolute_error(y_test, y_pred_svr)
mse_svr = mean_squared_error(y_test, y_pred_svr)
rmse_svr = mean_squared_error(y_test, y_pred_svr, squared=False)
r2_svr = r2_score(y_test, y_pred_svr)
rss_svr = np.sum((y_test - y_pred_svr) ** 2)

print(f"Support Vector Regressor (SVR):")
print(f"MAE: {mae_svr}")
print(f"MSE: {mse_svr}")
print(f"RSS: {rss_svr}")
print(f"RMSE: {rmse_svr}")
print(f"R^2: {r2_svr}")
```

# Metrics

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Example predictions from different models (replace with actual predictions)
y_pred_rf = rf_model.predict(X_test)  # Random Forest
y_pred_xgb = xgb_model.predict(X_test)  # XGBoost
y_pred_svr = svr_model.predict(X_test)  # SVR
y_pred_lr = lr_model.predict(X_test)  # Linear Regression

# Calculate metrics for each model
metrics = {
    'Model': ['Random Forest', 'XGBoost', 'SVR', 'Linear Regression'],
    'MAE': [
        mean_absolute_error(y_test, y_pred_rf),
        mean_absolute_error(y_test, y_pred_xgb),
        mean_absolute_error(y_test, y_pred_svr),
        mean_absolute_error(y_test, y_pred_lr)
    ],
    'MSE': [
        mean_squared_error(y_test, y_pred_rf),
        mean_squared_error(y_test, y_pred_xgb),
        mean_squared_error(y_test, y_pred_svr),
        mean_squared_error(y_test, y_pred_lr)
    ],
    'RMSE': [
        mean_squared_error(y_test, y_pred_rf, squared=False),
        mean_squared_error(y_test, y_pred_xgb, squared=False),
        mean_squared_error(y_test, y_pred_svr, squared=False),
        mean_squared_error(y_test, y_pred_lr, squared=False)
    ],
    'R²': [
        r2_score(y_test, y_pred_rf),
        r2_score(y_test, y_pred_xgb),
        r2_score(y_test, y_pred_svr),
        r2_score(y_test, y_pred_lr)
    ],
    'RSS': [
        np.sum((y_test - y_pred_rf) ** 2),  # RSS for Random Forest
        np.sum((y_test - y_pred_xgb) ** 2),  # RSS for XGBoost
        np.sum((y_test - y_pred_svr) ** 2),  # RSS for SVR
        np.sum((y_test - y_pred_lr) ** 2)  # RSS for Linear Regression
    ]
}

# Create DataFrame
metrics_df = pd.DataFrame(metrics)
metrics_df = metrics_df.round(2)
# Print the table
print(metrics_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics DataFrame
models = metrics_df['Model']
metrics = ['MAE', 'MSE', 'RMSE', 'R²', 'RSS']

# Set up the figure for subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

# Plot each metric
for i, metric in enumerate(metrics):
    axes[i].bar(models, metrics_df[metric], color=['skyblue', 'orange', 'green', 'red'])
    axes[i].set_title(f'Comparison of {metric}')
    axes[i].set_ylabel(metric)
    axes[i].set_xticks(np.arange(len(models)))
    axes[i].set_xticklabels(models, rotation=30, ha='right')
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Hide the last subplot (6th) if there are only 5 metrics
axes[-1].axis('off')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
